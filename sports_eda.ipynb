```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sports_data.csv")
df.head()
```

```python
dept_count = df['Department'].value_counts()
```

```python
dept_count
```

```python
plt.figure(figsize=(6,4))
dept_count.plot(kind='bar')
plt.title("Department-wise Sports Participation")
plt.xlabel("Department")
plt.ylabel("Number of Students")
plt.show()
```

- CSE department has the highest sports participation.
- IT department shows moderate participation.
- ME and EE departments have comparatively lower participation.

```python
sport_count = df['Sport'].value_counts()
sport_count
```

```python
plt.figure(figsize=(6,4))
sport_count.plot(kind='bar')
plt.title("Sport-wise Sports Participation")
plt.xlabel("Sport")
plt.ylabel("Number of Students")
plt.show()
```

- Cricket has the highest participation among students.
- Football shows moderate participation.
- Badminton has comparatively lower participation.


```python
corr_matrix = df[['Weekly_Hours', 'Matches_Played', 'Fitness_Score']].corr()
corr_matrix
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

- Weekly_Hours and Fitness_Score show a positive correlation.
- Matches_Played also has a positive impact on Fitness_Score.

- CSE department shows the highest participation in sports activities.
- Cricket is the most preferred sport among students.
- IT department shows moderate participation, while ME and EE departments have comparatively lower participation.
- Students who spend more weekly hours in sports tend to have higher fitness scores.
- There is a positive correlation between matches played and fitness level.


This project analyzed college sports participation data using Exploratory Data Analysis (EDA).
Department-wise and sport-wise analysis helped identify participation trends.
Correlation analysis showed that regular sports activity positively impacts students' fitness.
The study highlights the importance of promoting sports activities across all departments.
